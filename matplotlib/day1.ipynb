```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Opensourcefordatascience/Data-sets/master/blood_pressure.csv")

patient=df.head(10)["patient"]
bp_after=df.head(10)["bp_after"]
bp_before=df.head(10)["bp_before"]
```

```python
plt.xlabel("patient")
plt.ylabel("bp")
plt.plot(patient,bp_after, label ="after")
plt.plot(patient,bp_before)
plt.legend()
plt.grid()
```

```python
complany = ["Google" , "Amazon", "Meta", "NetFlix", "Apple"]
revenue= [100, 342, 200, 180, 280]
profit = [90, 100, 50,150,200]
```

```python
plt.bar(complany,revenue,width=0.5, label= "revenue")
plt.bar(complany,profit,width=0.5, label="profit")
plt.legend()
```

```python
plt.barh(complany,revenue, label= "revenue")
plt.barh(complany,profit,label="profit")
plt.legend()
```

```python
plt.pie(revenue,labels=complany,explode=[0,0,.2,0,0], autopct="%0.2f%%") 
```
